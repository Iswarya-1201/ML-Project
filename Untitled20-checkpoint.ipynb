```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("traffic.csv")
print("First 5 rows:\n", df.head())
```

```python
print("\nDataset Info:\n")
print(df.info())

print("\nMissing Values:\n", df.isnull().sum())

print("\nStatistical Summary:\n", df.describe())
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(numeric_df)
plt.show()
df.hist(figsize=(10,8))
plt.show()
```

```python
X = df[['Junction', 'Vehicles']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
df['Cluster'] = y_kmeans
```

```python
plt.figure(figsize=(8,6))

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_kmeans)

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='X')

plt.title("K-Means Clustering")
plt.xlabel("Junction (scaled)")
plt.ylabel("Vehicles (scaled)")
plt.show()
```

```python
print("\nCluster Means:\n")
print(df.groupby('Cluster').mean(numeric_only=True))
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y_kmeans)

plt.title("Clusters using PCA")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```
